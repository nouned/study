```python
#App: Study App
#Notes: I created this as the first assignment for a python coding course at General Assembly. It received full marks, exceeding expectations for the class.
#Author: Mike Stevenson
#Link: https://www.linkedin.com/in/mikestevensonmba/
# Â© Mike Stevenson 2023

#adds necessary packages for the program to work
import csv #for managing CSV file Question Bank
import random #for randomising order of questions
import os.path #used to checking question bank exists on initialisation

QUESTION_BANK_FILE = "Question-Bank.csv"

#create question mode
def create_question():
    question = input("Enter your study question: ")
    responses = input("Enter four possible responses, separated by commas: ").split(",")
    correct_response = input("Which response is correct (a, b, c, or d)? ")
    topic_tags = input("Enter topic tags, separated by commas: ").split(",")
    
    #writes the question inputs into the csv file 'Question-Bank'
    with open(QUESTION_BANK_FILE, mode="a", newline="") as csv_file:
        writer = csv.writer(csv_file)
        writer.writerow([question, responses, correct_response, topic_tags])
    
    # Ask user if they want to create another question, start a study session or quit
    while True:
        choice = input("Enter c to create another question, s to start a study session, or q to quit: ")
        if choice.lower() == "c":
            create_question()
        elif choice.lower() == "s":
            study_questions()
        elif choice.lower() == "q":
            print(f"Thanks for playing! See you again soon.")
            return
        else:
            print("Invalid choice. Please enter c to create another question, s to start a study session, or q to quit.")

# study mode        
def study_questions():
    with open(QUESTION_BANK_FILE) as csv_file:
        csv_reader = csv.reader(csv_file)
        questions = list(csv_reader)
        #defines what the program should do if there are currently no questions stored - prompt user to fill the question bank.
        if not questions:
            print("No questions found in the question bank. Please create a question first.")
            return
        #randomises the order of questions for the user to keep things interesting.
        random.shuffle(questions)
        for q in questions:
            print(q[0])
            responses = q[1].split(",")
            for i, r in enumerate(responses):
                print(f"{chr(97 + i)}) {r.strip('[]')}")  # remove brackets from each option
            answer = input("Enter your answer (a, b, c, or d): ")
            #makes user response lower case and verifies against lower case answer to avoid case sensitive errors.
            if answer.lower() == q[2].lower():
                print("Correct!")
            else:
                print(f"Your choice was incorrect. The correct answer is {q[2].upper()}.")
            choice = input("Enter y to continue or q to quit: ")
            #Comment to marker this assignment submission is taken from github
            while choice.lower() not in ["y", "q"]:  # check if input is valid
                print("Invalid choice. Please enter either y to continue or q to quit.")
                choice = input("Enter y to continue or q to quit: ")
            if choice.lower() == "q":
                print(f"Thanks for playing! See you again soon.")
                return
        print("Congratulations! You have answered all available questions.")
    
    # Ask user if they want to create another question, start a study session or quit
    while True:
        choice = input("Enter c to create a question, s to start another study session, or q to quit: ")
        if choice.lower() == "c":
            create_question()
        elif choice.lower() == "s":
            study_questions()
        elif choice.lower() == "q":
            print(f"Thanks for playing! See you again soon.")
            return
        else:
            print("Invalid choice. Please enter c to create a question, s to start another study session, or q to quit.")

#sets up main sequence where user chooses between create question and study mode, limits option to create mode if there are no questions yet.
def main():
    if not os.path.exists(QUESTION_BANK_FILE):
        print("No question bank found. Please create a question first.")
        create_question()
    else:
        choice = input("Enter c to create a question or s to study: ")
        if choice.lower() == "c":
            create_question()
        elif choice.lower() == "s":
            study_questions()
        else:
            print("Invalid choice, sorry. Please input either c to create a new practice question or s to start a study session")

if __name__ == "__main__":
    main()
```
